```python
import sys
import matplotlib.pyplot as plt
import numpy as np
from skimage import exposure
import cv2
from skimage import data
obraz = data.clock()

plt.figure(figsize=(10,10))
plt.imshow(obraz, cmap="gray")
plt.axis('on')
plt.show()

print(obraz.shape[0],obraz.shape[1])
    
```

# FUNKCJA

```python
def funkcja(image):
    dane=[]  #lista z danymi kolorów
    for i in range(256):   
        dane.append(0)
    
    for i in range(image.shape[0]):      #"Przjeście" przez cały obraz 
        for j in range (image.shape[1]):
            dane[image[i,j]]+=1          #uzupełnienie listy

    return dane
```

# Wywołanie funkcji i generowanie histogramu

```python
    cbins=[k for k in range(256)]  #zakres kolorów od 0 do 255
    plt.grid()                     #siatka wykresu
    plt.plot(cbins,funkcja(obraz))        
    plt.show()                     #wyświetlenie wykresu
    
    plt.grid() 
    plt.bar(cbins, funkcja(obraz))
    plt.show() 
    
    
```

# Histogram by MatPlot

```python
from skimage import exposure

histogram = exposure.histogram(obraz, nbins=32)
hist, cbins = histogram

plt.plot(cbins, hist)
plt.xlim([-10, 265])
plt.grid()
plt.show()

plt.bar(cbins, hist)
plt.xlim([-10, 265])
plt.grid()
plt.show()
```
