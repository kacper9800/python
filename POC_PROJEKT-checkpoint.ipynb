## Import potrzebnych bibliotek

# Definicja funkcji wy≈õwietlajƒÖcej dwa obrazy obok siebie

```python
def show2imgs(im1, im2, title1='Obraz pierwszy', title2='Obraz drugi', size=(10,10)):
    
    f, (ax1, ax2) = plt.subplots(1,2, figsize=size)
    ax1.imshow(im1, cmap='gray')
    ax1.axis('off')
    ax1.set_title(title1)

    ax2.imshow(im2, cmap='gray')
    ax2.axis('off')
    ax2.set_title(title2)
    plt.show()
```

# Binaryzacja

```python
th = filters.threshold_otsu(obraz)  #Automatyczny pr√≥g binaryzacji
print("Pr√≥g binaryzacji: ", th)     #Warto≈õƒá progu binaryzacji
th, bim = cv2.threshold(obraz, thresh=th, maxval=255, type=cv2.THRESH_BINARY_INV)   #Binaryzacja z inwersjƒÖ, bim - zmienna z obrazem

show2imgs(obraz, bim, title1='Obraz przed binaryzacja', title2='Obraz po binaryzacji', size=(10,10)) #Wy≈õwietlenie obraz√≥w
```

# Zamkniƒôcie obiekt√≥w

```python
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5))    #Maska
closing = cv2.morphologyEx(bim, cv2.MORPH_RECT, kernel)      #Operacja zamkniƒôcia z u≈ºyciem maski "kernel"

plt.imshow(closing)              #Nazwa obrazu na kt√≥rym pracujemy
plt.axis('on')
plt.suptitle('Obraz po przekszta≈Çceniach')
plt.show()
```

# Zliczanie obiekt√≥w

```python
label_objects, nb_labels = ndi.label(closing) #ka≈ºda unikatowa funkcja wej≈õcia ma unikalnƒÖ etykietƒô w zwr√≥conej tablicy.

sizes = np.bincount(label_objects.ravel())  #bincount liczy liczbƒô wystƒÖpie≈Ñ ka≈ºdej warto≈õci w tablicy nieujemnych int.
print(sizes)
mask_sizes = sizes > 20
mask_sizes[0] = 0
figures = mask_sizes[label_objects]

print("Liczba obiekt√≥w na obrazie:", nb_labels) 

plt.imshow(bim, cmap="autumn")
plt.axis('on')
plt.suptitle('Obiekty na obrazie')
plt.show()

#plt.imshow(label_objects,cmap='Accent')
#plt.show()
```

# Definicje funkcji

```python
no_of_object = 2

def cog(labelledImage, objNumber):
    cx = 0
    cy = 0
    for y in range(labelledImage.shape[0]):
        for x in range(labelledImage.shape[1]):
            if labelledImage[y,x] == objNumber:
                cy = cy + y
                cx = cx + x

    cx = cx / sizes[objNumber]
    cy = cy / sizes[objNumber]

    return int(cy), int(cx)

cogs = []
for i in range(nb_labels):
    cogs.append(cog(label_objects, i+1))

print(cogs)
```

# Rysowanie marker√≥w

```python
def drawMarkers(image, coords):

    image = np.copy(image)
    
    for item in coords:
        y,x = item
        for j in range(y-5, y+6):
            image[j,x] = 0
        for i in range(x-5, x+6):
            image[y,i] = 0
            
    return image


lo = drawMarkers(label_objects, cogs)
plt.figure(figsize=(10,10))
plt.imshow(lo,cmap='hot')
plt.show()
```

# Definicja funkcji obliczajƒÖcej wsp√≥≈Çczynnik Blaira-Blissa.

```python
def computeBB(points):
    
    s = len(points)
    my,mx = cog2(pts)
    
    r = 0
    for point in points:
         r = r + distance.euclidean(point,(my,mx))**2
            
    return s/(math.sqrt(2*math.pi*r))
```

# Definicja funkcji obliczajƒÖcej wsp√≥≈Çczynnik Fereta

```python
def computeFeret(points):
    
    px = [x for (y,x) in points]
    py = [y for (y,x) in points]
    
    fx = max(px) - min(px)
    fy = max(py) - min(py)
    
    return float(fy)/float(fx)    
```

# Liczba pikseli w obiekcie

```python
def getFigure(labelledImage, objNumber):
    
    points = []
    for y in range(labelledImage.shape[0]):
        for x in range(labelledImage.shape[1]):
            if labelledImage[y,x] == objNumber:
                points.append((y,x))

    return points
```

# ≈örodek ciƒô≈ºko≈õci

```python
def cog2(points):
    mx=0
    my=0
    for (y,x) in points:
        mx = mx + x
        my = my + y
    mx = mx/len(points)
    my = my/len(points)
    
    return [my, mx]
```

# Wy≈õwietlenie nastƒôpujƒÖcych informacji o obiektach:
 #### Liczba punkt√≥w
 #### ≈örodek ciƒô≈ºko≈õci
 #### Wsp√≥≈Çczynnik Blaira-Blissa
 #### Wsp√≥≈Çczynnik Fereta
 #### Procent jaki zajmujƒÖ obrazy w skali ca≈Çego obrazu

```python
suma =0
for i in range(nb_labels):
    pts = getFigure(label_objects, i+1)
    bb = computeBB(pts)
    feret = computeFeret(pts)
    suma = suma+ len(pts)
    
    print(i+1,'. Liczba punktow: ',len(pts),'\nSrodek ciezkosci: ', cog2(pts),'\nBlair-Bliss: ', bb,'\nFeret: ',feret, '\n+++++++++++++++++++++++++++++++++++++++++++++\n')

calosc = obraz.shape[0]*obraz.shape[1]
procent = (suma/calosc)*100
print('Liczba pikseli zajmowanych przez obiekty:',suma)
print('Liczba wszystkich pikseli:',calosc)
print('Obiekty zajmuja: ',procent,'% obrazu!')
```
